```python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Load pre-trained model and tokenizer
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Function to get sentence embedding
def get_embedding(sentence):
    inputs = tokenizer(sentence, return_tensors='pt', truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

# Sentences
reference_sentence = "The weather is cold today"
target_sentence = "It is freezing today"

# Get embeddings
reference_embedding = get_embedding(reference_sentence)
target_embedding = get_embedding(target_sentence)

# Compute cosine similarity
similarity = cosine_similarity([reference_embedding], [target_embedding])[0][0]

print("Cosine Similarity:", similarity)
```

The cosine similarity metric is 0.8663093
